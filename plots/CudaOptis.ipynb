## Time to solution of the CUDA implementations as a function of the number of pixels

```python
n_pxs = [2073600, 3686400, 8294400, 58982400, 92160000, 144000000, 207360000, 368640000]
opti3 = [1.70, 1.63, 2.06, 5.22, 7.31, 10.36, 14.46, 24.39]
opti4 = [1.05, 1.19, 1.25, 1.49, 1.58, 1.92, 2.26, 2.88]
opti5 = [1.14, 1.26, 1.15, 1.38, 1.58, 1.64, 1.91, 2.52]
all_optis = [opti3, opti4, opti5]
all_labels = ['opti3', 'opti4', 'opti5']
```

```python
plotting.plot_tts(n_pxs, all_optis, xscale='log', yscale='log', 
                  labels=all_labels, legend_title='CUDA implementation', 
                  xlabel='Number of pixels', ylabel='Time to solution [s]')
```

```python
plotting.plot_tts(n_pxs, all_optis, xscale='log', yscale='log', 
                  labels=all_labels, legend_title='CUDA implementation', 
                  xlabel='Number of pixels', ylabel='Time to solution [s]', 
                  filename='cuda-optis-px', saveas='svg')
```

## Time to solution of the serial implementations as a function of the number of pixels

```python
n_pxs = [2073600, 3686400, 8294400, 58982400, 92160000, 144000000, 207360000, 368640000]
naive = [25.30, 44.94, 100.00, 723.16, 1113.74, 1746.83, 2524.52, 4466.48]
opti1 = [12.98, 22.38, 49.90, 358.53, 557.68, 877.53, 1258.91, 2234.49]
opti2 = [1.32, 2.37, 5.32, 37.89, 58.41, 91.58, 132.53, 237.74]
opti3 = [1.29, 2.20, 5.08, 36.81, 56.85, 89.24, 128.57, 212.74]
all_optis = [naive, opti1, opti2, opti3]
all_labels = ['naive', 'opti1', 'opti2', 'opti3']
```

```python
plotting.plot_tts(n_pxs, all_optis, xscale='log', yscale='log', 
                  labels=all_labels, legend_title='Serial implementation', 
                  xlabel='Number of pixels', ylabel='Time to solution [s]')
```

```python
plotting.plot_tts(n_pxs, all_optis, xscale='log', yscale='log', 
                  labels=all_labels, legend_title='Serial implementation', 
                  xlabel='Number of pixels', ylabel='Time to solution [s]', 
                  filename='serial-optis-px', saveas='svg')
```
